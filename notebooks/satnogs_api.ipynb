```python
import requests
import pandas as pd
import time
import os
from dotenv import load_dotenv

load_dotenv()

# --- CONFIGURATION ---
# Replace with your actual API Token from https://db.satnogs.org/user/edit
SATNOGS_API_TOKEN=os.getenv('SATNOGS_API_TOKEN')

# NORAD ID of the satellite (e.g., 40012 is UPSat, 42784 is Pegasus)
# You can find these IDs on the satellite's page on db.satnogs.org
NORAD_ID = '40012' 

# Output filename
OUTPUT_FILE = f'scrap_data/satnogs_telemetry_{NORAD_ID}.json'
# ---------------------
```

```python
token = SATNOGS_API_TOKEN
satellite_id = NORAD_ID
page = 3

base_url = "https://db.satnogs.org/api/telemetry/"
headers = {
    'Authorization': f'Token {token}',
    'Accept': 'application/json',
}

params = {
    'satellite': satellite_id,
    'format': 'json',
    'page': page
}

all_data = []
```

```python
response = requests.get(base_url, headers=headers, params=params)
print(response.status_code)
```

```python
import json

with open(f"scrap_data/data_{satellite_id}_raw_p{page}.json", "w") as f:
    json.dump(response.json(),f, indent=4)
```
